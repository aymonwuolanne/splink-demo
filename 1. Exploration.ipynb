```python
import duckdb
from splink import DuckDBAPI
from splink.exploratory import profile_columns, completeness_chart
```

```python
con = duckdb.connect()

df_l = con.read_parquet("./perturbed_data_l.parquet")
df_r = con.read_parquet("./perturbed_data_r.parquet")

db_api = DuckDBAPI(connection=con)

df_l
```

```python
completeness_chart(
    [df_l, df_r],
    db_api,
    ['unique_id', 'fname', 'sname', 'sex', 'dob', 'sa4', 'address'],
    table_names_for_chart=['df_l', 'df_r'],
)
```

```python
profile_columns(
    [df_l, df_r],
    db_api,
)
```
